### Aggregate Data with the Python .agg() Function

working with the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
themes_by_year=sets.groupby('year').agg({'theme_id':pd.Series.nunique})
themes_by_year.rename(columns={'theme_id':'nr_themes'},inplace=True)
themes_by_year.head()
```

```python
set_by_years=sets.groupby('year').agg({'set_num':pd.Series.nunique})
set_by_years.rename(columns={'set_num':'un_setnum'},inplace=True)
set_by_years.head()
```

Ploting the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

```python
plt.plot(themes_by_year.index[:-2],themes_by_year['nr_themes'][:-2])
```

### Line Charts with Two Seperate Axes

```python
ax1=plt.gca()
ax2=ax1.twinx()

ax2.plot(themes_by_year.index[:-2],themes_by_year['nr_themes'][:-2],color='b')
ax1.plot(set_by_year.index[:-2],set_by_year['un_setnum'][:-2],color='g')
ax1.set_xlabel('year')
ax1.set_ylabel('numberof parts')
ax2.set_ylabel('number of themes')
```

Using <code>.groupby()</code> and <code>.agg()</code> function together to figure out the average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

```python
parts_per_set=sets.groupby('year').agg({'num_parts':pd.Series.mean})
parts_per_set.rename(columns={'num_parts':'avg_parts'},inplace=True)
parts_per_set.head()
```

```python
parts_per_set.tail()
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
set_theme_count=pd.DataFrame({'id':set_theme_count.index,'set_count':set_theme_count.values})
set_theme_count.head()
```

```python
merge_df=pd.merge(set_theme_count,themes,on='id')
merge_df.head()
```
